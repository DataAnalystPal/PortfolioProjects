## <font color = '#0CB371'> Importing </font> Libraries

```python
## fundamental libraries
import numpy as np
import pandas as pd

## visualizrion libraries
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

## scikit-learn libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
## supressing warnings
import warnings
warnings.filterwarnings('ignore')
```

## <font color = '#0CB371'> Importing </font> Dataset

```python
## importing dataset from directory and
## loading it in dataframe
data = pd.read_csv('housing.csv')
```

```python
## verifying the imported dataset
data.head()
```

```python
## basic info
data.info()
```

```python
## dropping the missing values
data.dropna(inplace = True)
```

```python
## basic info
data.info()
```

## <font color = '#0CB371'> Splitting </font> Dataset

```python
x = data.drop(['median_house_value'], axis = 1)
y = data['median_house_value']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
training_data = x_train.join(y_train)
training_data
```

## <font color = '#0CB371'> Exploring  </font> Dataset and Processing

```python
training_data.hist(figsize = (15, 10))

plt.show()
```

```python
corr_mat = training_data.corr(numeric_only =True)
corr_mat
```

```python
plt.figure(figsize = (15, 8))
sns.heatmap(corr_mat, annot = True, cmap = 'YlGnBu')
plt.show()
```

```python
training_data
```

```python
training_data['total_rooms'] = np.log(training_data['total_rooms'] + 1)
training_data['total_bedrooms'] = np.log(training_data['total_bedrooms'] + 1)
training_data['population'] = np.log(training_data['population'] + 1)
training_data['households'] = np.log(training_data['households'] + 1)
```

```python
training_data.hist(figsize = (15, 8))
plt.show()
```

```python
## one-hot encoding
training_data.ocean_proximity.value_counts()
```

```python
training_data = training_data.join(pd.get_dummies(training_data['ocean_proximity'], dtype = 'int')).drop(['ocean_proximity'], axis = 1)
```

```python
training_data
```

```python
plt.figure(figsize = (15, 8))
sns.scatterplot(x = 'latitude', y = 'longitude', data = training_data, hue = 'median_house_value', palette = 'coolwarm')

plt.show()
```

## <font color = '#0CB371'> Feature </font> Engineering

```python
## feature engineering

training_data['bedroom_ratio'] = training_data['total_bedrooms'] / training_data['total_rooms']
training_data['household_rooms'] = training_data['total_rooms'] / training_data['households']
```

## <font color = '#0CB371'> Linear </font> Regression

```python
x_train, y_train = training_data.drop(['median_house_value'], axis = 1), training_data['median_house_value']
```

```python
reg = LinearRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
test_data = x_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

test_data = test_data.join(pd.get_dummies(test_data['ocean_proximity'], dtype = 'int')).drop(['ocean_proximity'], axis = 1)

test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms'] / test_data['households']

x_test, y_test = test_data.drop(['median_house_value'], axis = 1), test_data['median_house_value']
```

```python
x_train
```

```python
reg.score(x_test, y_test)
```

```python
y_test = list(y_test)
y_test
```

```python
## Plotting
plt.figure(figsize = (15, 8))

plt.plot(y_test, color = 'red', label = 'Actual Values')
plt.plot(test_data, color = 'green', label = 'Predicted Values')

plt.title('Actual Values vs. Predicted Values')
plt.xlabel('Number of Values')
plt.ylabel('Median House Value ')


plt.show()
```
